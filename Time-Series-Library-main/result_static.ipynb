```python
# 文件夹结果统计
import csv
import re
import os

models = ['Autoformer', 'DLinear', 'iTransformer', 'Nonstationary_Transformer', 'PatchTST', 'TimesNet']

# 正则表达式模式，用于匹配 MSE 和 MAE
metrics_pattern = re.compile(r'mse:\s*(nan|[0-9.e+-]+),\s*mae:\s*(nan|[0-9.e+-]+),\s*dtw:-999')

dataset_pattern = re.compile(r'Data Path:\s*(.+?)_all_roll_ewma_span28_normalize.csv*')


# 获取 logs 文件夹下以 top 开头的文件
log_files = [f for f in os.listdir('logs/aseaRandomTop100')]
print(log_files)

# CSV 输出路径
csv_output_path = 'workresult/ASEA_Random_top100/Result.csv'

# 检查文件是否为空
file_is_empty = False
try:
    with open(csv_output_path, 'r', newline='') as csv_file:
        file_is_empty = csv_file.read().strip() == ''
except FileNotFoundError:
    file_is_empty = True

# 打开 CSV 文件，并处理日志文件
with open(csv_output_path, 'a', newline='') as csv_file:
    writer = csv.writer(csv_file)
    if file_is_empty:
        writer.writerow(['dataset', 'mse', 'mae'])

    # 处理每个日志文件
    for log_file_name in log_files:
        # 初始化数据集名称
        dataset_name = None

        # 构造完整路径并打开日志文件
        log_file_path = os.path.join('logs/aseaRandomTop100', log_file_name)
        try:
            with open(log_file_path, 'r') as log_file:
                for line in log_file:
                    # 匹配数据集名称
                    dataset_match = dataset_pattern.search(line)
                    if dataset_match:
                        dataset_name = dataset_match.group(1)

                    # 匹配 MSE 和 MAE
                    metrics_match = metrics_pattern.search(line)
                    if metrics_match:
                        mse, mae = metrics_match.groups()
                        # 将 MSE 和 MAE 保留到小数点后三位
                        try:
                            mse_val = f"{float(mse):.3f}" if mse != 'nan' else mse
                        except ValueError:
                            mse_val = 'nan'
                        try:
                            mae_val = f"{float(mae):.3f}" if mae != 'nan' else mae
                        except ValueError:
                            mae_val = 'nan'

                        # 写入 CSV 文件
                        if dataset_name:
                            writer.writerow([dataset_name, mse_val, mae_val])
                            dataset_name = None
                        else:
                            print("Warning: Found metrics without a dataset name.")
        except FileNotFoundError:
            print(f"Warning: Log file not found - {log_file_path}")
            continue

print(f"Results written to {csv_output_path}")
```
