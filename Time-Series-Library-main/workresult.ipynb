```python
# 读文件夹算每个csv文件的均值和整体均值
import pandas as pd
import os

def calculate_mse_means_from_folder(folder_path):
    """
    读取指定文件夹及其子文件夹下的所有CSV文件，并计算每个文件中mse列的均值，
    最后计算所有文件mse均值的平均值。
    
    参数:
    folder_path (str): 包含CSV文件的文件夹路径
    """
    # 检查文件夹是否存在
    if not os.path.exists(folder_path):
        print(f"错误：文件夹 {folder_path} 不存在")
        return
    
    # 获取文件夹中所有CSV文件（包括子文件夹）
    csv_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.csv'):
                # 保存相对路径（相对于folder_path）
                relative_path = os.path.relpath(os.path.join(root, file), folder_path)
                csv_files.append(relative_path)
    
    if not csv_files:
        print(f"在 {folder_path} 及其子文件夹中未找到CSV文件")
        return
    
    # 存储所有文件的mse均值
    mse_means = []
    
    # 遍历所有CSV文件并计算mse均值
    for relative_file_path in csv_files:
        file_path = os.path.join(folder_path, relative_file_path)
        try:
            # 读取CSV文件
            df = pd.read_csv(file_path)
            
            # 检查是否存在mse列
            if 'mse' not in df.columns:
                print(f"警告：文件 {relative_file_path} 中未找到mse列")
                continue
            
            # 计算mse列的均值
            mse_mean = df['mse'].mean()
            
            # 存储mse均值用于后续计算总体平均
            mse_means.append(mse_mean)
            
            # 输出文件相对路径和mse均值
            print(f"文件 {relative_file_path} 中 mse 列的均值为: {mse_mean}")
            
        except Exception as e:
            print(f"处理文件 {relative_file_path} 时出错: {e}")
    
    # 计算所有文件mse均值的平均值
    if mse_means:
        overall_mean = sum(mse_means) / len(mse_means)
        print(f"\n所有文件的 MSE 均值的平均值为: {overall_mean}")
    else:
        print("没有成功计算任何文件的 MSE 均值")

# 使用示例
folder_path = 'workresult/stacking/generalization_baseline'  # 替换为你的文件夹路径
calculate_mse_means_from_folder(folder_path)

# folder_path = 'workresult/stacking/Parameter_Sensitivity_Analysis/d_model'  # 替换为你的文件夹路径
# calculate_mse_means_from_folder(folder_path)


# folder_path = 'workresult/stacking/Parameter_Sensitivity_Analysis/n_heads'  # 替换为你的文件夹路径
# calculate_mse_means_from_folder(folder_path)
```

```python
# 参数敏感性分析
import pandas as pd
import os
import re

def extract_parameter_from_filename(filename):
    """
    从文件名中提取参数名和参数值
    例如: exp_attention_fusion_noHistoricalEncoder_noResidual_2model_PatchTST_Dlinear_100dataset_dmodel_64.csv
    提取: ('dmodel', '64')
    """
    # 移除文件扩展名
    filename = os.path.splitext(filename)[0]
    
    # 定义可能的参数模式
    parameter_patterns = [
        r'd_model_(\d+)',
        r'dmodel_(\d+)',
        r'nheads_(\d+)',
        r'n_heads_(\d+)',
        r'd_ff_(\d+)',
        r'dff_(\d+)',
        r'patch_len_(\d+)',
        r'stride_(\d+)',
        r'e_layers_(\d+)',
        r'd_layers_(\d+)'
    ]
    
    # 尝试匹配各种参数模式
    for pattern in parameter_patterns:
        match = re.search(pattern, filename)
        if match:
            param_name = pattern.split('_')[0].replace('(', '').replace(')', '')
            param_value = match.group(1)
            return param_name, param_value
    
    # 如果没有找到标准参数，尝试查找最后的参数
    parts = filename.split('_')
    for i in range(len(parts)-1, 0, -1):
        if parts[i].isdigit() and not parts[i-1].isdigit():
            param_name = parts[i-1]
            param_value = parts[i]
            return param_name, param_value
    
    return None, None

def parameter_sensitivity_analysis(folder_path):
    """
    对指定文件夹中的CSV文件进行参数敏感性分析
    
    参数:
    folder_path (str): 包含CSV文件的文件夹路径
    """
    # 检查文件夹是否存在
    if not os.path.exists(folder_path):
        print(f"错误：文件夹 {folder_path} 不存在")
        return
    
    # 存储结果
    results = {}
    
    # 获取文件夹中所有CSV文件
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    if not csv_files:
        print(f"在 {folder_path} 中未找到CSV文件")
        return
    
    # 遍历所有CSV文件
    for csv_file in csv_files:
        file_path = os.path.join(folder_path, csv_file)
        try:
            # 提取参数信息
            param_name, param_value = extract_parameter_from_filename(csv_file)
            
            if param_name is None or param_value is None:
                print(f"警告：无法从文件名 {csv_file} 中提取参数信息")
                continue
            
            # 读取CSV文件
            df = pd.read_csv(file_path)
            
            # 检查是否存在mse列
            if 'mse' not in df.columns:
                print(f"警告：文件 {csv_file} 中未找到mse列")
                continue
            
            # 计算mse列的均值
            mse_mean = df['mse'].mean()
            
            # 按参数名分组存储结果
            if param_name not in results:
                results[param_name] = []
            
            results[param_name].append({
                'parameter_value': param_value,
                'mse_mean': mse_mean,
                'filename': csv_file
            })
            
            print(f"文件 {csv_file} 中参数 {param_name}={param_value} 的MSE均值为: {mse_mean}")
            
        except Exception as e:
            print(f"处理文件 {csv_file} 时出错: {e}")
    
    # 确保输出文件夹存在
    output_folder = "workresult/stacking/Parameter_Sensitivity_Analysis"
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
        print(f"已创建文件夹: {output_folder}")
    
    # 为每个参数生成单独的结果文件
    for param_name, data_list in results.items():
        # 创建DataFrame
        result_df = pd.DataFrame(data_list)
        
        # 按参数值排序
        result_df = result_df.sort_values('parameter_value')
        
        # 保存到CSV文件
        output_filename = f"{output_folder}/{param_name}.csv"
        result_df.to_csv(output_filename, index=False)
        print(f"\n参数 {param_name} 的敏感性分析结果已保存到: {output_filename}")
        
        # 打印该参数的统计信息
        print(f"\n参数 {param_name} 的统计信息:")
        print(result_df.to_string(index=False))

# 使用示例
folder_path = 'workresult/stacking/Parameter_Sensitivity_Analysis/d_model'  # 替换为你的文件夹路径
parameter_sensitivity_analysis(folder_path)
```

```python
# 单一参数敏感性分析 画图
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

def plot_parameter_sensitivity_from_csv(csv_file_path):
    """
    根据指定的CSV文件路径绘制参数敏感性分析图表
    
    参数:
    csv_file_path (str): 参数敏感性分析结果CSV文件的路径
    """
    # 检查文件是否存在
    if not os.path.exists(csv_file_path):
        print(f"错误：文件 {csv_file_path} 不存在")
        return
    
    try:
        # 读取CSV文件
        df = pd.read_csv(csv_file_path)
        
        # 获取参数名称（从文件名）
        param_name = os.path.splitext(os.path.basename(csv_file_path))[0]
        
        # 确保数据类型正确
        df['parameter_value'] = pd.to_numeric(df['parameter_value'])
        df['mse_mean'] = pd.to_numeric(df['mse_mean'])
        
        # 按参数值排序
        df = df.sort_values('parameter_value')
        
        # 创建图表（使用默认大小）
        plt.figure()
        
        # 实现x轴均分显示
        x_indices = range(len(df))
        plt.plot(x_indices, df['mse_mean'], marker='o', linewidth=2, markersize=8)
        
        # 添加标题和标签
        plt.title(f'"{param_name}" Sensitivity_Analysis', fontsize=16, fontweight='bold')
        plt.xlabel('param_value', fontsize=12)
        plt.ylabel('MSE mean', fontsize=12)
        
        # 添加网格
        plt.grid(True, alpha=0.3)
        
        # 设置x轴刻度标签
        plt.xticks(x_indices, df['parameter_value'])
        plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
        
        # 调整布局
        plt.tight_layout()
        
        # 显示图表
        plt.show()
        
        print(f"参数 {param_name} 的敏感性分析图表绘制完成")
        
    except Exception as e:
        print(f"处理文件 {csv_file_path} 时出错: {e}")

def plot_multiple_parameters_from_csv(csv_files_list):
    """
    根据多个CSV文件路径绘制参数敏感性分析对比图表
    
    参数:
    csv_files_list (list): 参数敏感性分析结果CSV文件路径列表
    """
    # 检查文件是否存在
    existing_files = [f for f in csv_files_list if os.path.exists(f)]
    if not existing_files:
        print("错误：没有找到任何指定的文件")
        return
    
    # 创建一个包含所有参数的综合图表（使用默认大小）
    plt.figure()
    
    # 为每个参数创建一条线
    for csv_file_path in existing_files:
        try:
            # 读取CSV文件
            df = pd.read_csv(csv_file_path)
            
            # 获取参数名称（从文件名）
            param_name = os.path.splitext(os.path.basename(csv_file_path))[0]
            
            # 确保数据类型正确
            df['parameter_value'] = pd.to_numeric(df['parameter_value'])
            df['mse_mean'] = pd.to_numeric(df['mse_mean'])
            
            # 按参数值排序
            df = df.sort_values('parameter_value')
            
            # 实现x轴均分显示
            x_indices = range(len(df))
            
            # 绘制线图
            plt.plot(x_indices, df['mse_mean'], marker='o', linewidth=2, 
                     markersize=6, label=param_name)
            
        except Exception as e:
            print(f"处理文件 {csv_file_path} 时出错: {e}")
            continue
    
    # 添加标题和标签
    plt.title('参数敏感性分析对比', fontsize=16, fontweight='bold')
    plt.xlabel('参数值', fontsize=12)
    plt.ylabel('MSE 均值', fontsize=12)
    
    # 添加图例
    plt.legend()
    
    # 添加网格
    plt.grid(True, alpha=0.3)
    
    # 调整布局
    plt.tight_layout()
    
    # 显示图表
    plt.show()
    
    print("多参数敏感性分析对比图表绘制完成")

# 使用示例
# 绘制单个参数的敏感性分析图表
plot_parameter_sensitivity_from_csv("workresult/stacking/Parameter_Sensitivity_Analysis/nheads.csv")
plot_parameter_sensitivity_from_csv("workresult/stacking/Parameter_Sensitivity_Analysis/dmodel.csv")

# 绘制多个参数的对比图表
# csv_files = [
#     "workresult/stacking/Parameter_Sensitivity_Analysis/nheads.csv",
#     "workresult/stacking/Parameter_Sensitivity_Analysis/dmodel.csv"
# ]
# plot_multiple_parameters_from_csv(csv_files)
```

```python
# 温度敏感性分析-文件梳理
import pandas as pd
import os
import re

def analyze_temperature_results(base_folder_path="workresult/stacking/temperature"):
    """
    分析不同温度下模型组合的MSE表现，并生成汇总CSV文件
    
    参数:
    base_folder_path (str): 包含温度子文件夹的根路径
    """
    # 检查根文件夹是否存在
    if not os.path.exists(base_folder_path):
        print(f"错误：文件夹 {base_folder_path} 不存在")
        return
    
    # 获取所有温度子文件夹
    temperature_folders = [f for f in os.listdir(base_folder_path) 
                          if os.path.isdir(os.path.join(base_folder_path, f)) and f.startswith('temperature')]
    
    if not temperature_folders:
        print(f"在 {base_folder_path} 中未找到温度子文件夹")
        return
    
    # 提取温度值并排序
    temperatures = []
    for folder in temperature_folders:
        match = re.search(r'temperature([\d.]+)', folder)
        if match:
            temperatures.append((float(match.group(1)), folder))
    
    # 按温度值排序
    temperatures.sort(key=lambda x: x[0])
    
    if not temperatures:
        print("未找到有效的温度文件夹")
        return
    
    print("找到以下温度设置:")
    for temp, folder in temperatures:
        print(f"  温度 {temp}: {folder}")
    
    # 收集所有模型组合名称
    model_combinations = set()
    
    # 读取所有温度下的数据
    temperature_data = {}
    
    for temp_value, folder_name in temperatures:
        folder_path = os.path.join(base_folder_path, folder_name)
        csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
        
        if not csv_files:
            print(f"在 {folder_path} 中未找到CSV文件")
            continue
        
        temp_results = {}
        for csv_file in csv_files:
            file_path = os.path.join(folder_path, csv_file)
            try:
                # 从文件名提取模型组合名（移除温度相关信息）
                model_name = re.sub(r'_temperature[\d.]+', '', os.path.splitext(csv_file)[0])
                model_combinations.add(model_name)
                
                # 读取CSV文件并计算MSE均值
                df = pd.read_csv(file_path)
                if 'mse' not in df.columns:
                    print(f"警告：文件 {csv_file} 中未找到mse列")
                    continue
                
                mse_mean = df['mse'].mean()
                temp_results[model_name] = mse_mean
                
            except Exception as e:
                print(f"处理文件 {csv_file} 时出错: {e}")
                continue
        
        temperature_data[temp_value] = temp_results
    
    if not model_combinations:
        print("未找到任何模型组合")
        return
    
    # 创建结果DataFrame
    results_data = []
    
    for model_name in sorted(model_combinations):
        row_data = {'model_combination': model_name}
        
        # 为每个温度添加MSE均值
        for temp_value in sorted(temperature_data.keys()):
            mse_value = temperature_data[temp_value].get(model_name, None)
            row_data[f'temp_{temp_value}'] = mse_value
        
        results_data.append(row_data)
    
    # 创建DataFrame并保存到CSV
    results_df = pd.DataFrame(results_data)
    
    # 保存结果到CSV文件
    output_file = os.path.join(base_folder_path, "temperature_analysis_results.csv")
    results_df.to_csv(output_file, index=False)
    
    print(f"\n分析完成！结果已保存到: {output_file}")
    print("\n结果预览:")
    print(results_df.to_string(index=False))

# 运行分析
analyze_temperature_results()
```

```python
# 算error correlation均值
import pandas as pd
import os

def calculate_corr_means_from_folder(folder_path):
    """
    读取指定文件夹下的所有CSV文件，并计算每个文件中列名包含"Corr"的列的均值，
    最后计算所有文件相关列均值的平均值。
    
    参数:
    folder_path (str): 包含CSV文件的文件夹路径
    """
    # 检查文件夹是否存在
    if not os.path.exists(folder_path):
        print(f"错误：文件夹 {folder_path} 不存在")
        return
    
    # 获取文件夹中所有CSV文件
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    if not csv_files:
        print(f"在 {folder_path} 中未找到CSV文件")
        return
    
    # 存储所有文件的相关列均值
    corr_means = []
    
    # 遍历所有CSV文件并计算包含"Corr"的列的均值
    for csv_file in csv_files:
        file_path = os.path.join(folder_path, csv_file)
        try:
            # 读取CSV文件
            df = pd.read_csv(file_path)
            
            # 查找列名中包含"Corr"的列（但列名不等于"Corr"）
            corr_columns = [col for col in df.columns if 'Corr' in col and col != 'Corr']
            
            if not corr_columns:
                print(f"警告：文件 {csv_file} 中未找到列名包含'Corr'且不等于'Corr'的列")
                continue
            
            # 计算每个包含"Corr"的列的均值
            file_corr_means = {}
            for col in corr_columns:
                col_mean = df[col].mean()
                file_corr_means[col] = col_mean
                corr_means.append(col_mean)
            
            # 输出文件名和各相关列的均值
            print(f"文件 {csv_file} 中包含'Corr'的列:")
            for col, mean_val in file_corr_means.items():
                print(f"  {col} 列的均值为: {mean_val}")
            
        except Exception as e:
            print(f"处理文件 {csv_file} 时出错: {e}")
    
    # 计算所有相关列均值的平均值
    if corr_means:
        overall_mean = sum(corr_means) / len(corr_means)
        print(f"\n所有包含'Corr'列的均值的平均值为: {overall_mean}")
    else:
        print("没有成功计算任何包含'Corr'列的均值")

# 使用示例
folder_path = 'workresult/stacking/correlation'  # 替换为你的文件夹路径
calculate_corr_means_from_folder(folder_path)
```

```python
def calculate_mse_improvement_ratio(mse_before, mse_after):
    """
    计算 MSE 提升比例
    
    参数:
    mse_before (float): 提升前的 MSE 值
    mse_after (float): 提升后的 MSE 值
    
    返回:
    float: 改良比例 (%)
    """
    if mse_before <= 0:
        raise ValueError("MSE 提升前的值必须大于 0")
    
    improvement = (mse_before - mse_after) / mse_before * 100
    return improvement

mse_before = 4.66 #单patchtst
mse_after = 4.40
ratio = calculate_mse_improvement_ratio(mse_before, mse_after)
print(f"MSE 改良比例: {ratio:.2f}%")
```

```python
from datetime import datetime

def calculate_duration(start_time_str, end_time_str, time_format="%H.%M.%S"):
    """
    计算两个时间点之间的持续时间
    
    参数:
    start_time_str (str): 起始时间字符串，格式需与 time_format 一致
    end_time_str (str): 结束时间字符串，格式需与 time_format 一致
    time_format (str): 时间字符串的解析格式，默认为 "%H.%M.%S"
    
    返回:
    timedelta: 持续时间（以小时、分钟和秒表示）
    float: 总持续时间（以秒为单位）
    """
    # 将字符串转换为 datetime 对象
    start_time = datetime.strptime(start_time_str, time_format)
    end_time = datetime.strptime(end_time_str, time_format)

    # 计算时间差
    duration = end_time - start_time

    # 输出持续时间（以秒为单位）
    total_seconds = duration.total_seconds()
    
    return duration, total_seconds

# 示例使用
# stacking下单一modify_patchtst模型的训练时间
start_time = "9.06.12"
end_time = "9.15.08"
duration, total_seconds = calculate_duration(start_time, end_time)

print(f"持续时间: {duration}")
print(f"总持续时间（秒）: {total_seconds:.2f} 秒")

#stacking下，modify_patchtst和dlinear运行时间
start_time = "15.32.10"
end_time = "15.44.49"
duration, total_seconds = calculate_duration(start_time, end_time)

print(f"持续时间: {duration}")
print(f"总持续时间（秒）: {total_seconds:.2f} 秒")
```
